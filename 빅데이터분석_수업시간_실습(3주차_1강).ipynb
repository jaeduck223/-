```python
pd.read_csv('ex6.csv',nrows=4) #nrow 옵셔을 사용해서 읽어오는 양을 지정(head 기능과 유사함.)
```

```python
#파일을 여러조각으로 나누어 읽고 싶다면, chunksize 옵션으로 row의 갯수를 지정

chunker = pd.read_csv('ex6.csv', chunksize=1000)
chunker
```

```python
#ex6.csv파일을 chunker사이즈로 읽어 순회하며 'key'row에 있는 값을 세어보는 예

tot = pd.Series([], dtype='int64') #비어있는 시리즈를 형성함

for piece in chunker:
    tot = tot.add(piece['key'].value_counts(), fill_value=0)
tot = tot.sort_values(ascending=False)
tot[:10] ###중요함!!!!!! #######################################
```

```python
data.to_csv('out.csv')  #to_csv 메소드를 사용하면 데이터를 쉽표로 구분된 형식으로 파일에 쓸 수 았다.
```

```python
#다른 구분자도 가능 : 쉼표 대신 |를 구분자로 사용한 예
import sys

data.to_csv(sys.stdout, sep=':') #콘솔에서 확인할 수 있도록 실제 파일에 기록하지 않고 stdout에 결과를 기록
```

```python
data.to_csv(sys.stdout, na_rep='널체크') #누락된 값은 비어있는 문자열로 나타나는데, 이것 역시 원하는 값으로 지정가능하다.
```

```python
#row와 col 이름 없이 출력

data.to_csv(sys.stdout, index=False, header=False)
```

```python
#row와 col의 일부만 기록할 수 있다.

data.to_csv(sys.stdout, index=False, columns=['a', 'b', 'c'])
```
